```python
import pygimli as pg
import pygimli.meshtools as mt
```

```python
left = -30
right = 30
depth = 25
```

```python
world = mt.createWorld(start = [left,0], end = [right, -depth], layers = [-5])
pg.show(world);
```

```python
line = mt.createLine(start=[left,-20], end=[right,-15])
pg.show(line);
```

```python
geometry = world + line
pg.show(geometry);
```

```python
body = mt.createPolygon([[-8, -12], [0, -7], [8, -13]], isClosed=True, marker = 3, addNodes = 5, interpolate = 'spline')
pg.show(body, showNodes = True);

geometry += body
pg.show(geometry, boundaryMarkers = True);
```

```python
print(geometry)
```

```python
mesh = mt.createMesh(geometry, quality = 34, area = 1, smooth = True)
print(mesh)
pg.show(mesh);
```

```python
mesh.save('mesh.bms')
```
